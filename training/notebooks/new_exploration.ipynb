# Merge the datasets

```python
import pandas as pd

p1 = pd.read_csv('/Users/dwibon/Desktop/ISL-FingerSpell/dataset/abhigyan_dataset.csv')
p2 = pd.read_csv('/Users/dwibon/Desktop/ISL-FingerSpell/dataset/tanisha_dataset.csv')
p3 = pd.read_csv('/Users/dwibon/Desktop/ISL-FingerSpell/dataset/abid_dataset.csv')
p4 = pd.read_csv('/Users/dwibon/Desktop/ISL-FingerSpell/dataset/mriganka_dataset.csv')
p5 = pd.read_csv('/Users/dwibon/Desktop/ISL-FingerSpell/dataset/nayan_dataset.csv')
p6 = pd.read_csv('/Users/dwibon/Desktop/ISL-FingerSpell/dataset/nilotpal_dataset.csv')

combined_v2 = pd.concat(
    [p1, p2, p3, p4, p5, p6],
    ignore_index=True
)

combined_v2 = combined_v2.dropna()
combined_v2.to_csv('../../dataset/combined_train_v2.csv', index=False)

print(f"p1: {p1.shape}")
print(f"p2: {p2.shape}")
print(f"p3: {p3.shape}")
print(f"p4: {p4.shape}")
print(f"p5: {p5.shape}")
print(f"p6: {p6.shape}")
print(f"\nCombined v2: {combined_v2.shape}")
print(f"\nSamples per sign:\n{combined_v2['label'].value_counts().sort_index()}")
```
